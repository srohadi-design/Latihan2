```python
import pandas as pd
```

```python
data_london_df = pd.read_csv('/Users/user/dataset/london_crime_by_lsoa.csv')
```

```python
data_london_df
```

```python
data_london_df2 = data_london_df.copy()
```

```python
data_london_total = data_london_df2.groupby('borough',axis=0)['value'].sum()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use(['ggplot'])
```

```python
data_london_total.plot(kind='pie',figsize=(14,8),autopct='%1.1f%%',labels=None)
plt.title('Total Crime')
plt.axis('equal')
plt.legend(labels=data_london_total.index,loc='upper left')
plt.show()
```
